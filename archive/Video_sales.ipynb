```python
#importing dataset
import pandas as pd
df= pd.read_csv("vgsales.csv")
df
```

```python
#calculating the null values in the data set
df.isna().sum()
```

```python
#dropping the values of null values
df=df.dropna(subset=['Publisher','Year'],axis=0)
df=df.reset_index(drop=True)
df.isna().sum()
```

```python
# dropping ALL duplicate values
df.drop_duplicates(subset ="Name",
                     keep = False, inplace = True)
df
```

```python
#to know the data types present in the dataset
df.info()
```

```python
df["Year"]=df["Year"].astype(int)
df['Year'].dtype
```

```python
df.head(10)
```

```python
#selecting the variables.
X=df[['NA_Sales', 'EU_Sales']]
X
y=df['Global_Sales']
y
```

```python


X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
#importing neccessary modules
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
#building a Linear Regression model

lr = LinearRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_pred
```

```python
#building a Decession Tree model
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
dtr_pred = dtr.predict(X_test)
dtr_pred
```
